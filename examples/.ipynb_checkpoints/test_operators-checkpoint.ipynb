```python
import dsatools
import dsatools.utilits as ut
from dsatools.generator import harmonics
signal1 = harmonics(amplitude=[3,0.3,1],
                    f0=[0.01,0.1, 1,2,1.4,2.1],
                    length=512,                    
                    delta_f=[0.2,0.1,0.3],snr_db=25)
ut.probe(signal1)
x = signal1
ut.probe(dsatools.operators.correlation(x, mode='full',take_mean=False,unbias=False,))
```

```python
import dsatools
import dsatools.utilits as ut
from dsatools.generator import harmonics
signal1 = harmonics(amplitude=[3,0.3,1],
                    f0=[0.01,0.1, 1,2,1.4,2.1],
                    length=512,                    
                    delta_f=[0.2,0.1,0.3],snr_db=25)
ut.probe(signal1)
x = signal1
df = 0.1
signal2 = harmonics(amplitude=[3,0.3,1],
                    f0=[0.01,0.1, 1,2,1.4,2.1],
                    length=512,                    
                    delta_f=[0.2+df,0.1,0.3],snr_db=25)
y = signal2
R_x,R_y = dsatools.operators.xycor(x, y, mode='R12',modecor='same')
ut.probe(R_x)
R_x,R_y = dsatools.operators.xycor(x, y, mode='Rfb12',modecor='same')
ut.probe(R_x)
```

```python
import dsatools
import dsatools.utilits as ut
from dsatools.generator import harmonics
import matplotlib.pyplot as plt
signal1 = harmonics(amplitude=[3,0.3,1],
                    f0=[0.01,0.1, 1,2,1.4,2.1],
                    length=512,                    
                    delta_f=[0.2,0.1,0.3],snr_db=25)
ut.probe(signal1)
x = signal1

xcdf, bnins = dsatools.operators.ecdf(x)
plt.plot(bnins,xcdf);plt.show()
xhist, bnins = dsatools.operators.hist(x,n_bins = 50)
plt.plot(bnins,xhist)
```

```python
import dsatools
import numpy as np
import matplotlib.pyplot as plt
from dsatools import utilits as ut

s1=dsatools.generator.harmonics(f0=[2,2.9,0.8],delta_f=[0,0,0.1],fs=20,length=125).real
s2=dsatools.generator.harmonics(f0=[2,3.1,0.7],fs=20,length=125,phase0=[np.pi/3,np.pi/5]).real
ut.probe(s2)
ut.probe(s1)
print(ut.euclidian_matrix(s1,s2))
```

```python
import dsatools
import dsatools.utilits as ut
from dsatools.generator import harmonics

import matplotlib.pyplot as plt
signal1 = harmonics(amplitude=[3,0.3,1],
                    f0=[0.01,0.1, 1,2,1.4,2.1],
                    length=512,                    
                    delta_f=[0.2,0.1,0.3],snr_db=25)
ut.probe(signal1)
x = ut.hilbert(signal1.real)
ut.probe(x)
x = ut.hilbert_autocorrelation(signal1.real)
ut.probe(x)
```
