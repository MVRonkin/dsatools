```python
import numpy as np
import matplotlib.pyplot as plt
import dsatools
from dsatools import roots
from dsatools import utilits as ut
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[0.4,1,1.05,2,2.1],
                                        length=512,                    
                                        delta_f=[0.2],)

ut.probe(signal1)
x = signal1
fs=10

freqs = roots.music(x,order=6,mode='full', fs=10)
print('music \t',np.sort(freqs))
freqs = roots.ev(x,order=6,mode='full', fs=10)
print('ev \t',np.sort(freqs))
freqs = roots.phd(x,order=6,mode='full',fs=10)
print('phd \t',np.sort(freqs))
freqs = roots.phd_cor(x,order=6,mode='covar', cor_mode='same', fs=10)
print('phd_cor \t',np.sort(freqs))
freqs = roots.minvariance(x,order=6,mode='full')
print('[minvariance \t',np.sort(freqs))
freqs = roots.minnorm(x,order=6,mode='full',signal_space=False, fs=10)
print('minnorm \t',np.sort(freqs))
freqs = roots.hoyw(x,order=6,mode='full')
print('hoyw \t',np.sort(freqs))
freqs = roots.esprit(x,order=6, mode='full', fs=10)
print('esprit \t',np.sort(freqs))
freqs = roots.esprit_cor(x,order=6, mode='covar',cor_mode='same', fs=10)
print('esprit_cor \t',np.sort(freqs))
freqs = roots.matrix_pencil(x,order=6, mode='full', fs=10)
print('matrix_pencil \t',np.sort(freqs))
freqs = roots.matrix_pencil_cov(x,order=6, mode='full', fs=10)
print('matrix_pencil_cov \t',np.sort(freqs))
freqs = roots.matrix_pencil_cor(x,order=6, mode='covar',cor_mode='same', fs=10)
print('matrix_pencil_cor \t',np.sort(freqs))
freqs = roots.kernel_noisespace(x,order=6, mode='full',kernel='rbf',kpar=0.0002, fs=10)
print('kernel_noisespace \t',np.sort(freqs))
freqs = roots.kernel_signalspace(x,order=6, mode='full',kernel='rbf',kpar=0.0002, fs=10)
print('kernel_signalspace \t',np.sort(freqs))
freqs = roots.kernel_minnorm(x,order=6, mode='full',kernel='rbf',kpar=0.0002,signal_space=False, fs=10)
print('kernel_minnorm \t',np.sort(freqs))
freqs = roots.kernel_minnorm(x,order=6, mode='full',kernel='rbf',kpar=0.0002,signal_space=True, fs=10)
print('kernel_minnorm \t',np.sort(freqs))
freqs = roots.kernel_esprit(x,order=6, mode='full',kernel='rbf',kpar=0.0002, fs=10)
print('kernel_esprit \t',np.sort(freqs))
freqs = roots.kernel_matrix_pencil(x,order=6, mode='full',kernel='rbf',kpar=0.0002, fs=10)
print('kernel_matrix_pencil \t',np.sort(freqs))
```
