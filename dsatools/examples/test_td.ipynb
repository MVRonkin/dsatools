# PCA

```python
import dsatools
from dsatools import operators
from dsatools import utilits as ut
from dsatools._base._time_domain_decomposition import *

signal1 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,2,3,3.1,2.04],
                    delta_f=[0.4],fs=10)
ut.probe(signal1)
x  = signal1

psd = pca(signal1, order=25, mode='covar')
ut.probe((psd.sum(axis=0)))
ut.probe(psd[0])

psd = pca(signal1, order=25, mode='full')
ut.probe((psd.sum(axis=0)))

psd = pca(signal1, order=35, mode='toeplitz')
ut.probe((psd.sum(axis=0)))
```
