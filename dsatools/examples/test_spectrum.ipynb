```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import spectrum
from dsatools import utilits as ut
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[0.4,1,1.05,2,2.1],
                                        length=512,                    
                                        delta_f=[0.2],)

ut.probe(signal1); plt.show()
x = signal1

psd = spectrum.music(x,order=6,mode='full')
plt.plot(psd);plt.title('music');plt.show()
psd = spectrum.ev(x,order=6,mode='full')
plt.plot(psd);plt.title('ev');plt.show()
psd = spectrum.phd(x,order=50,mode='full')
plt.plot(psd);plt.title('phd');plt.show()
psd = spectrum.phd_cor(x,order=50,mode='covar', cor_mode='same')
plt.plot(psd);plt.title('phd_cor');plt.show()
psd = spectrum.minvariance(x,order=60,mode='full')
plt.plot(psd);plt.title('minvariance');plt.show()
psd = spectrum.minnorm(x,order=16,mode='full',signal_space=False)
plt.plot(psd);plt.title('minnorm');plt.show()
psd = spectrum.hoyw(x,order=162,mode='full')
plt.plot(psd);plt.title('hoyw');plt.show()
psd = spectrum.kernel_noisespace(x,order=6, mode='full',kernel='rbf',kpar=0.0002,)
plt.plot(psd);plt.title('kernel_noisespace');plt.show()
psd = spectrum.kernel_signalspace(x,order=6, mode='full',kernel='rbf',kpar=0.0002,)
plt.plot(psd);plt.title('kernel_signalspace');plt.show()
psd = spectrum.kernel_minnorm(x,order=46, mode='full',kernel='rbf',kpar=0.0002,signal_space=False)
plt.plot(psd);plt.title('kernel_minnorm_noise');plt.show()
psd = spectrum.kernel_minnorm(x,order=46, mode='full',kernel='rbf',kpar=0.0002,signal_space=True)
plt.plot(psd);plt.title('kernel_minnorm_signal');plt.show()
```

```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import spectrum
from dsatools import utilits as ut
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[0.4,1,1.05,2,2.1],
                                        length=512,                    
                                        delta_f=[0.2],)

ut.probe(signal1); plt.show()
x = signal1

psd = spectrum.correlogram(x)
plt.plot(psd); plt.title('correlogram');plt.show()
psd =  spectrum.periodogram(x, window=('dpss',0.1),x_range=[0,100])
plt.plot(psd); plt.title('periodogram'); plt.show()
psd =  spectrum.bartlett(x, n_sections=3)
plt.plot(psd); plt.title('bartlett'); plt.show()
psd =  spectrum.daniell(x,n_average=3,)
plt.plot(psd); plt.title('daniell'); plt.show()
psd =  spectrum.welch(x,window='hann',window_length=500,step=10,)
plt.plot(psd); plt.title('welch'); plt.show()
psd = spectrum.blackman_tukey(x,mode='full',window='parzen',)
plt.plot(psd); plt.title('blackman_tukey'); plt.show()
psd = spectrum.kernel_periodogram(x, mode='covar',kernel='poly',kpar=1.05,window='hann',)
plt.plot(psd); plt.title('kernel_periodogram'); plt.show()
```

```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import spectrum
from dsatools import utilits as ut
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[0.4,1,1.05,2,2.1],
                                        length=512,                    
                                        delta_f=[0.2],)

ut.probe(signal1); plt.show()
x = signal1

psd = spectrum.capone(x, order=15, mode='full')
plt.plot(psd); plt.title('capone'); plt.show()
psd = spectrum.slepian(x, order=15)
plt.plot(psd); plt.title('slepian'); plt.show()

x = dsatools.operators.correlation(signal1,mode='full')
psd = spectrum.capone(x, order=15, mode='full')
plt.plot(psd); plt.title('capone+cor'); plt.show()
psd = spectrum.slepian(x, order=15)
plt.plot(psd); plt.title('slepian+cor'); plt.show()
```
