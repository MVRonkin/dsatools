# matrix_pencil

## matrix_pencil

```python
import dsatools
from dsatools import utilits as ut
from dsatools._base._polyroot_decomposition import *

signal1 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,2,3,3.1,2.04],
                    delta_f=[0.4],fs=10)

ut.probe(signal1)
x = signal1

order = 30
roots = matrix_pencil(x, order, mode = 'covar' ,tls_rank = None)
psds = roots2decomposition(x,roots=roots,order = order)
ut.probe(psds.sum(axis=0))

signal0 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,0.2,3],
                    delta_f=[0.0],fs=10)
ut.probe(signal0)
x = signal0
order = 6
roots = matrix_pencil(x, order, mode = 'covar' ,tls_rank = None)
freqs = roots2freqs(roots=roots,order = order, fs = 10)
print(freqs)
```
