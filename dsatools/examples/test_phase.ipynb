```python
import dsatools
from dsatools import operators
from dsatools import utilits as ut
from dsatools.phase import *

fs = 10
signal0 = dsatools.generator.harmonics(amplitude=[1],
                                      f0      = [1],
                                      delta_f = [0.0],
                                      phase0  = [0],
                                      fs=fs,
                                      snr_db=30,)
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                      f0=[1],
                                      delta_f=[0.0],
                                      phase0 = [0.4],
                                      fs=fs,
                                      snr_db=30,)
ut.probe(signal0)
x  = signal0
print(wls_phase(signal0,signal1))
print(ls_phase(signal0,signal1))
print(tretter_phase(signal0,signal1))
print(maxcor_phase(signal0,signal1,normalize = False))
print(maxcor_phase(signal0,signal1,normalize = True))
```
