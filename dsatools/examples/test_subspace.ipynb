# MUSIC

```python
import dsatools
from dsatools import operators
from dsatools import utilits as ut
import matplotlib.pyplot as plt
from dsatools._base._subspace import *

signal1 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,2,3,3.1,2.04],
                    delta_f=[0.4],fs=10)
signal0 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,2,3],
                    delta_f=[0.0],fs=10)

ut.probe(signal1); plt.show()
ut.probe(signal0); plt.show()

order = 15
x = signal1
noise_space = music(signal1, order, mode='full', lags=None)
psd = subspace2psd(noise_space )
plt.plot(psd); plt.show()

order = 3
x = signal0
noise_space = music(signal0, order, mode='full', lags=None)
freqs = subspace2freq(noise_space,order )
print(freqs)

```
