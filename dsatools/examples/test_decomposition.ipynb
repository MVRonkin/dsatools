```python
import dsatools
from dsatools import operators
import dsatools.utilits as ut
from dsatools import decomposition
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[0.4,1,1.05,2,2.1],
                                        length=512,                    
                                        delta_f=[0.2],)

ut.probe(signal1)
x = signal1

psds = decomposition.dmd(operators.correlation(x,mode='full'),order=26,mode='covar')
ut.probe(psds.sum(axis=0),title='dmd+cor')
psds = decomposition.pca_cor(x,order=26,mode='full',cor_mode='full')
ut.probe(psds.sum(axis=0),title='pca_cor')
psds = decomposition.ssa(operators.correlation(x,mode='same'),order=26,mode='covar')
ut.probe(psds.sum(axis=0),title='ssa+cor')
psds = decomposition.esprit_cor(x,order=6,mode='covar',cor_mode='same')
ut.probe(psds.sum(axis=0),title='esprit_cor')
psds = decomposition.matrix_pencil_cor(x,order=36,mode='full',cor_mode='same')
ut.probe(psds.sum(axis=0),title='matrix_pencil_cor')
```

```python
import dsatools
from dsatools import operators
import dsatools.utilits as ut
from dsatools import decomposition
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[0.4,1,1.05,2,2.1],
                                        length=512,                    
                                        delta_f=[0.2],)

ut.probe(signal1)
x = signal1
psds = decomposition.hvd(x,order=6, fpar=30)
ut.probe(psds.sum(axis=0),title='hvd')
ut.probe(psds[0],title='hvd_zero')

psds = decomposition.ewt(x,order=6, gamma=0.2, average=4)
ut.probe(psds.sum(axis=0),title='ewt')
ut.probe(psds[0],title='ewt_zero')

psds = decomposition.vmd(x,order=6, reularization=2000,noise_impact=0,max_iter=50,tolerance=1e-03,)
ut.probe(psds.sum(axis=0),title='vmd')
ut.probe(psds[0],title='vmd_zero')
```

```python
import dsatools
from dsatools import operators
import dsatools.utilits as ut
from dsatools import decomposition
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[0.4,1,1.05,2,2.1],
                                        length=512,                    
                                        delta_f=[0.2],)

ut.probe(signal1)
x = signal1
psds = decomposition.emd(x,order=6, method='cubic', max_itter=100, tol=0.1)
ut.probe(psds.sum(axis=0),title='emd_cubic')
ut.probe(psds[0],title='emd_zero')
psds = decomposition.emd(x,order=6, method='rbf', max_itter=100, tol=0.1)
ut.probe(psds.sum(axis=0),title='emd_rbf')
ut.probe(psds[0],title='emd_zero')
psds = decomposition.emd(x,order=6, method='thin_plate', max_itter=100, tol=0.1)
ut.probe(psds.sum(axis=0),title='emd_thin_plate')
ut.probe(psds[0],title='emd_zero')
psds = decomposition.emd(x,order=6, method='akim', max_itter=100, tol=0.1)
ut.probe(psds.sum(axis=0),title='emd_akim')
ut.probe(psds[0],title='emd_zero')
psds = decomposition.emd(x,order=6, method='linear', max_itter=100, tol=0.1)
ut.probe(psds.sum(axis=0),title='emd_linear')
ut.probe(psds[0],title='emd_zero')
psds = decomposition.emd(x,order=6, method='cubic', max_itter=1000, tol=0.01)
ut.probe(psds.sum(axis=0),title='emd_cubic')
ut.probe(psds[0],title='emd_zero')
```
