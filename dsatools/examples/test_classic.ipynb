# Unparametric

# periodograms 

```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import utilits as ut
from dsatools._base._classic_psd import * 
signal1 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,2,3,3.1,2.04],
                    delta_f=[0.4],fs=10)
x = signal1
ut.probe(x); plt.show()

psd = correlogram(x)
plt.plot(psd); plt.show()
psd = periodogram(x, window=('dpss',0.1),x_range=[0,100])
plt.plot(psd); plt.show()
psd = bartlett(x, n_sections=3)
plt.plot(psd); plt.show()
psd = daniell(x,n_average=3,)
plt.plot(psd); plt.show()
psd = welch(x,window='hann',window_length=500,step=10,)
plt.plot(psd); plt.show()
psd = welch(x,window=None,window_length=500,step=10,)
plt.plot(psd); plt.show()
```

## blackman_tukey

```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import utilits as ut
from dsatools._base._classic_psd import * 
signal1 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,2,3,3.1,2.04],
                    delta_f=[0.4],fs=10)
x = signal1
ut.probe(x); plt.show()

psd = blackman_tukey(x,mode='full',window=None,)
plt.plot(psd); plt.show()
psd = blackman_tukey(x,mode='full',window='parzen',)
plt.plot(psd); plt.show()
psd = blackman_tukey(x,mode='toeplitz',window='parzen',)
plt.plot(psd); plt.show()
```

## kernel_periodogram

```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import utilits as ut
from dsatools._base._classic_psd import * 
signal1 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[1,2,3,3.1,2.04],
                    delta_f=[0.4],fs=10)
x = signal1
ut.probe(x); plt.show()

psd = kernel_periodogram(x, mode='full',kernel='linear',kpar=1,window=None,)
plt.plot(psd); plt.show()
psd = kernel_periodogram(x, mode='full',kernel='poly',kpar=1.05,window=None,)
plt.plot(psd); plt.show()
psd = kernel_periodogram(x, mode='covar',kernel='poly',kpar=1.05,window=None,)
plt.plot(psd); plt.show()
psd = kernel_periodogram(x, mode='covar',kernel='poly',kpar=1.05,window=None,)
plt.plot(psd); plt.show()
psd = kernel_periodogram(x, mode='covar',kernel='poly',kpar=1.05,window='hann',)
plt.plot(psd); plt.show()
```
