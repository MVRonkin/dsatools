```python
import dsatools
from dsatools import operators
import dsatools.utilits as ut
signal1 = dsatools.generator.harmonics(amplitude=[1],
                                        f0=[1.1,2],
                                        length=512,                    
                                        delta_f=[0.0],)

ut.probe(signal1)
x = signal1
```

```python
import dsatools
from dsatools import operators,spectrum
import dsatools.utilits as ut

psd = spectrum.music(x,order=4,mode='full')
plt.plot(psd)
```

```python
import dsatools
from dsatools import operators,spectrum
import dsatools.utilits as ut


psds = dsatools.decomposition.dmd(x,order=4,mode='full')
ut.probe(2*psds.sum(axis=0))
ut.probe(psds[0])
```

```python
import dsatools
from dsatools import operators,spectrum
import dsatools.utilits as ut

freqs = dsatools.roots.esprit(x,order=4,mode='full', fs=1)
print(freqs)
```

```python
import dsatools
from dsatools import operators,spectrum
import dsatools.utilits as ut

a,err = dsatools.arma.ar_burg_covar(x, order=4, mode='straight', unbias=False)
print(dsatools.arma.ar2freq(a,order=4,fs=1))
```

```python
import dsatools
from dsatools import operators,spectrum
import dsatools.utilits as ut

freqs = dsatools.frequency.fitz_r(x, fs=1)
print(freqs)
```
