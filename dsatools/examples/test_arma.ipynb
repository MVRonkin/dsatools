# ar_burg

```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import utilits as ut
from dsatools.arma import *

signal1 = dsatools.generator.harmonics(amplitude=[1],f0=[1,2,3,3.1,2.04],delta_f=[0.4],fs=10)

ut.probe(signal1);plt.show()

a,err=ar_burg(signal1,order=15)
ut.probe(arma2psd(a,b=1,err=1,n_psd=signal1.size), title='arma2psd')
ut.probe(arma2impresponce(a,b=1, n_samples=signal1.size), title='impulse responce')
    
signal0 = dsatools.generator.harmonics(amplitude=[1],f0=[1,2,3],delta_f=[0.0],fs=10)
a,err=ar_burg(signal0,order=3)
ut.probe(ar2predict(signal0,a,n_predict=2*signal0.size), title='ar 2 predict')
print('ar2freq',ar2freq(a,fs=10))
ut.probe(ar2decomposition(signal0,a,n_psd=signal0.size).sum(axis=0), title='ar 2 decomposition')
```

```python
import matplotlib.pyplot as plt
import dsatools
from dsatools import utilits as ut
from dsatools import operators
from dsatools.arma import *

signal1 = dsatools.generator.harmonics(amplitude=[1],f0=[1,2,3,3.1,2.04],delta_f=[0.4],fs=10)

ut.probe(signal1);plt.show()



poles_order = 55
zeros_order = 5
a,b,err=arma_prony(signal1, poles_order, zeros_order, mode='full')

ut.probe(arma2psd(a,b=b,err=1,n_psd=signal1.size), title='arma2psd')
ut.probe(arma2impresponce(a,b=b, n_samples=signal1.size), title='impulse responce')
    
signal0 = dsatools.generator.harmonics(amplitude=[1],f0=[1,2,3],delta_f=[0.0],fs=10)
poles_order = 6
zeros_order = 6
order = 6
a,b,err=arma_prony(signal0, poles_order, zeros_order, mode='full')
ut.probe(ar2predict(signal0,a,n_predict=2*signal0.size), title='ar 2 predict')
print('ar2freq',ar2freq(a,order,fs=10))
ut.probe(ar2decomposition(signal0,a,n_psd=signal0.size).sum(axis=0), title='ar 2 decomposition')
```
